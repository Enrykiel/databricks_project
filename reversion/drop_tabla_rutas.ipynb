### External location

```python
%sql
DROP EXTERNAL LOCATION IF EXISTS `extl-metastore` FORCE;
DROP EXTERNAL LOCATION IF EXISTS `extl-raw` FORCE;
DROP EXTERNAL LOCATION IF EXISTS `extl-bronze` FORCE;
DROP EXTERNAL LOCATION IF EXISTS `extl-silver` FORCE;
DROP EXTERNAL LOCATION IF EXISTS `extl-golden` FORCE;
```

### Catalogo

```python
%sql
DROP CATALOG IF EXISTS catalog_dev CASCADE;
```

```python
%sql
DROP SCHEMA IF EXISTS catalog_dev.raw;
DROP SCHEMA IF EXISTS catalog_dev.bronze;
DROP SCHEMA IF EXISTS catalog_dev.silver;
DROP SCHEMA IF EXISTS catalog_dev.golden;
```
