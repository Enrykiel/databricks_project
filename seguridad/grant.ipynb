### Rol de Data Engineers (DEs)

```python
%sql
-- CATALOGO
GRANT ALL PRIVILEGES ON CATALOG catalog_dev TO `DEs`;

-- PERMISOS ESQUEMAS
GRANT CREATE, USE SCHEMA ON SCHEMA  catalog_dev.bronze to `DEs`;
GRANT CREATE, USE SCHEMA ON SCHEMA  catalog_dev.silver to `DEs`;
GRANT CREATE, USE SCHEMA ON SCHEMA  catalog_dev.golden to `DEs`;

-- PERMISOS TABLAS
GRANT SELECT ON TABLE catalog_dev.bronze.ciclo TO `DEs`;
GRANT SELECT ON TABLE catalog_dev.bronze.matricula TO `DEs`;
GRANT SELECT ON TABLE catalog_dev.bronze.programa TO `DEs`;

GRANT SELECT ON TABLE catalog_dev.silver.matriculados_unicos_escuela TO `DEs`;
GRANT SELECT ON TABLE catalog_dev.silver.programa_estatus_calif TO `DEs`;

GRANT SELECT ON TABLE catalog_dev.golden.cant_matriculados_unicos_escuela_salud TO `DEs`;
GRANT SELECT ON TABLE catalog_dev.golden.programa_por_cant_estatus_calif TO `DEs`
```

```python
%sql
-- PERMISOS EXTERNAL LOCATION
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION `extl-metastore` to `DEs`;
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION `extl-bronze` to `DEs`;
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION `extl-silver` to `DEs`;
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION `extl-golden` to `DEs`;
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION `extl-raw` to `DEs`;
```
