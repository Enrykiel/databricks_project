```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import functions as F
```

```python
dbutils.widgets.text("catalogo", "catalog_dev")
dbutils.widgets.text("esquema_source", "silver")
dbutils.widgets.text("esquema_sink", "golden")
```

```python
catalogo = dbutils.widgets.get("catalogo")
esquema_source = dbutils.widgets.get("esquema_source")
esquema_sink = dbutils.widgets.get("esquema_sink")
```

### transform_mat_escuela

```python
df_transform_mat_escuela = spark.table(f"{catalogo}.{esquema_source}.matriculados_unicos_escuela")
```

```python
df_transform_mat_escuela.write.mode("overwrite").insertInto(f"{catalogo}.{esquema_sink}.cant_matriculados_unicos_escuela_salud")
```
